**Extraction des informations pertinentes pour l'entrainement de nos modèles :**

D'après l'analyse de la nouvelle filtred_bids il se voit qu'on peut extraire des variables qui semble importantes lors du developpement du modèle: 
- Une variable **total_bids** pour stocker le nombre de transactions totale effectué par le même bidder. Cette caractéristique capture l'activité globale d'un bidder.
- Une variable **avg_time_between_bids** la moyenne du temps entre les transactions. Cette caractéristique peut indiquer le comportement de soumission des enchères d'un bidder, s'il est actif de manière régulière ou sporadique.
- Une variable **most_frequent_merchandise** la catégorie fréquemment choisit par le bidder. Cela peut révéler les préférences d'un bidder en termes de types d'objets auxquels il participe.
- Une variable **most_frequent_country** le pays le plus fréquent.
- Une variable **unique_auctions** Le nombre d'enchères uniques auxquelles un bidder a participé. Cela donne un aperçu de la diversité des enchères auxquelles un bidder participe.
- Une variable **unique_device_ratio** le ratio du nombre de devices uniques utilisés par un bidder par rapport au nombre total de bids. Cela peut donner des informations sur la variété des devices utilisés par un bidder.


```python
# Calcul du nombre de transactions total efféctué par un meme bidder
total_bids= filtred_bids.groupby('bidder_id')['bid_id'].count().reset_index()
total_bids.columns =['bidder_id','total_bids']

# Calcul du temps moyens entre les différentes transaction
avg_time_between_bids=filtred_bids.groupby('bidder_id')['time'].diff().groupby(filtred_bids['bidder_id']).mean().reset_index()
avg_time_between_bids.columns =['bidder_id','avg_time_between_bids']

# Calcul de la catégorie fréquement choisit
most_frequent_merchandise= filtred_bids.groupby('bidder_id')['merchandise'].agg(lambda x: x.value_counts().index[0] if not 
                                                                                x.empty else None).reset_index()
most_frequent_merchandise.columns =['bidder_id','most_frequent_merchandise']

# Calcul du Pays frequement utilisé par le bidder
most_frequent_country= filtred_bids.groupby('bidder_id')['country'].agg(lambda x: x.value_counts().index[0] if not x.empty 
                                                                        and len(x.value_counts()) > 0 else None).reset_index()
most_frequent_country.columns =['bidder_id','most_frequent_country']

# Calcul du nombre d'enchère unique
unique_auctions=filtred_bids.groupby('bidder_id')['auction'].nunique().reset_index()
unique_auctions.columns =['bidder_id','unique_auctions']

#Calcul du ratio des devices unique par rapport au total devices utilisé

unique_device_ratio= filtred_bids.groupby('bidder_id')['device'].nunique()/filtred_bids.groupby('bidder_id')['bid_id'].count()
unique_device_ratio=unique_device_ratio.reset_index()
unique_device_ratio.columns = ['bidder_id', 'unique_device_ratio']

```

- Maintenant, c'est le temps de construire la nouvelle trainset en combinant les différent variables

```python
unique_device_ratio
```

```python
#Fusion des différentes variables en un Dataframe unique
new_features=pd.merge(total_bids,unique_auctions,on='bidder_id',how='left')
new_features=pd.merge(new_features,avg_time_between_bids,on='bidder_id',how='left')
new_features=pd.merge(new_features,most_frequent_merchandise,on='bidder_id',how='left')
new_features=pd.merge(new_features,most_frequent_country,on='bidder_id',how='left')
new_features=pd.merge(new_features,unique_device_ratio,on='bidder_id',how='left')

#Remplaçement des NAN
new_features=new_features.fillna(0)
print(new_features.isna().sum())
```

```python
#Affichage du nouveau Dataframe
new_features
```

- Création du nouveau train set :

```python
train = pd.merge(train,new_features,on='bidder_id',how='left')
train = train.fillna(0)
```
